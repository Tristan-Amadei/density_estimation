# <font color=darkcyan> Estimation de densités</font> 

#### <font color=darkcyan>Application à la détection de dauphins dans le Golfe de Gascogne</font>

Dans ce projet on s'intéresse à l'estimation d'une densité de probabilité $p_{\star}$ sur un espace $\mathsf{X} \subset \mathbb{R}^d$. On suppose que l'on dispose de variables $(X_i)_{1\leqslant i\leqslant n}$ indépendantes et identiquement distribuées et telles que la loi de $X_1$ admet la densité $p_\star$ par rapport à une mesure de référence sur $\mathsf{X}$.

Nous allons proposer, étudier et implémenter différents estimateurs de $p_\star$ à partir des $(X_i)_{1\leqslant i\leqslant n}$.

- **Estimateur à base d'histogrammes**.
- **Estimateurs à noyaux**.

Notons $\widehat{p}_n$ un estimateur de $p_\star$. La performance de cet estimateur est en générale mesurée à l'aide de deux critères.

``L'erreur quadratique moyenne`` est définie, pour tout $x\in \mathsf{X}$ par 

$$
\mathcal{E}(x) = \mathbb{E}\left[\left(\widehat{p}_n(x)-p_{\star}(x)\right)^2\right]\,.
$$

``L'erreur quadratique moyenne intégrée`` est définie par 

$$
\overline{\mathcal{E}} = \mathbb{E}\left[\int_0^1\left(\widehat{p}_n(x)-p_{\star}(x)\right)^2 dx\right]\,.
$$


#### <font color=darkred>Estimateus à base d'histogrammes</font>

Dans cette section on suppose que $\mathsf{X}$ est borné et on suppose que $\mathsf{X}=[0,1)$ sans perte de généralités. Pour tout $m\geqslant 1$ on considère $(A_1,\ldots, A_m)$ la partition uniforme de $[0,1)$ définie, pour tout $1\leqslant i \leqslant m$, par $A_i = [(i-1/m),i/m)$. Soient $(X_1\ldots,X_n)$ des variables aléatoires i.i.d. de loi de densité $p_{\star}$ par rapport à la mesure de Lebesgue sur $[0,1)$. On considère l'estimateur par histogramme de la densité $p_{\star}$ :

$$
\widehat{p}^h_n:x\mapsto \frac{1}{nh}\sum_{i=1}^n\sum_{j=1}^m 1_{A_j}(X_i)1_{A_j}(x)\,,
$$

où $h=1/m$.

### Erreur quadratique moyenne

$$\mathcal{E}(x) = \sum_{j=1}^m 1_{A_j}(x).\frac{P(X_i \in A_j)(1 - P(X_i \in A_j))}{nh^2} + \left( \sum_{j=1}^m 1_{A_j}(x).\frac{P(X_i \in A_j)}{h} - p_{\star}(x) \right)^2\$$

### Erreur quadratique moyenne intégrée

$$\overline{\mathcal{E}} = \int_0^1 p^2_{\star}(x) dx + \sum_{j=1}^m\frac{P(X_i \in A_j)(1 - P(X_i \in A_j))}{n.m} - \sum_{j=1}^m \frac{p^2_j}{h} $$

<font color=white> **S2** Ecrire une fonction produisant l'estimateur pour une valeur de $m$ donnée. Tracer l'estimateur pour différentes valeurs de $m$.</font>

````python
def hist_estimate(data, m, x):
    """
    Inputs
    ----------
    data: np.array contenant les X_i.
    m: nombre d'éléments de la partition.
    
    Outputs
    -------
    hist_est: fonction retournant l'estimation de la densité pour chaque x donné en argument.
    """
    n = data_delphins.shape[0]
    hist_est = 0
    for i in range(n):
        X_i = data[i]
        for j in range(m):
            if (j-1)/m <= X_i and X_i < j/m and (j-1)/m <= x and x < j/m:
                hist_est += 1
    hist_est *= m/n
    return hist_est
````

#### <font color=darkred>Estimateurs à base de noyaux</font>

La densité $p_{\star}$ est estimée par l'estimateur à noyau suivant :

$$
\widehat{p}^{h_n}_n:x\mapsto \frac{1}{nh_n}\sum_{i=1}^n K\left(\frac{X_i-x}{h_n}\right)\,.
$$

où $K:\mathbb{R}\to \mathbb{R}$ est une fonction positive telle que $\int_{\mathbb{R}}K(x) dx = 1$ et où $h_n$ est un réel strictement positif.

$\text{Soit x}\in \mathsf{X}.$
$$
Var(\widehat{p}^{h_n}_n(x)) = \frac{1}{nh^2_n} Var\left(K\left(\frac{X_1 - x}{h_n}\right)\right)  \text{par indépendance et loi i.d} 
\newline = \frac{1}{nh^2_n}  \mathbb{E}\left[\left(K\left(\frac{X_1 - x}{h_n}\right) - \mathbb{E}\left[K\left(\frac{X_1 - x}{h_n}\right) \right]\right)^2\right] 
 \newline \leq \frac{1}{nh^2_n} \mathbb{E} \left[K\left(\frac{X_1 - x}{h_n}\right)^2\right]
 \newline \leq \frac{1}{nh^2_n} \int_{\mathbb{R}} K\left(\frac{z - x}{h_n}\right) ^2 dz
\newline \leq \frac{1}{nh^2_n} \int_{\mathbb{R}} K\left(u\right) ^2.h_n du \text{  , changement de variable: u = } \frac{z-x}{h_n}
\newline \leq C_1.(nh_n)^{-1} \text{ où } C_1 = \int_{\mathbb{R}} K\left(u\right)^2 du, C_1 < \infty
$$

$\text{Soit x}\in \mathsf{X}.$
$$
biais(\widehat{p}^{h_n}_n(x)) = \mathbb{E} \left[\widehat{p}^{h_n}_n(x) - p_\star(x) \right]
\newline \mathbb{E} \left[\widehat{p}^{h_n}_n(x) \right] = \frac{1}{h_n} \int_{\mathbb{R}} K\left(\frac{z - x}{h_n}\right) p_\star(z) dz
\newline = \int_{\mathbb{R}} K(u) p_\star(u.h_n+x) du \text{  ,u = } \frac{z-x}{h_n}$$
$\newline \text{On a donc: }$
$$
biais(\widehat{p}^{h_n}_n(x)) = \int_{\mathbb{R}} K(u) p_\star(u.h_n+x) - p_\star(x) du
\newline = \int_{\mathbb{R}} K(u) (p_\star(u.h_n+x) - p_\star(x)) du \text{ car} \int_{\mathbb{R}} K(u) = 1$$ $\newline$
$\newline p_\star \text{ est deux fois dérivable alors, par Taylor-Young, pour tout u} \in \mathbb{R} \text{ il existe } \xi^{u,x}_n \in [x, x+u.h_n] \text{ tel que:}$ $\newline$
$$\newline p_\star(u.h_n+x) - p_\star(x) = uh_n.p_\star'(uh_n) + \frac{(uh_n)^2}{2} p_\star''(\xi^{u,x}_n)
$$
$\newline \text{D'où: }$
$$
biais(\widehat{p}^{h_n}_n(x)) = \int_{\mathbb{R}} K(u) uh_n.p_\star'(uh_n) du + \int_{\mathbb{R}} K(u)  \frac{(uh_n)^2}{2} p_\star''(\xi^{u,x}_n)du
\newline \int_{\mathbb{R}}u.K(u) du = 0 \implies \int_{\mathbb{R}} K(u) uh_n.p_\star'(uh_n) du = 0 \text{ d'où}
\newline biais(\widehat{p}^{h_n}_n(x)) = \int_{\mathbb{R}} K(u)  \frac{(uh_n)^2}{2} p_\star''(\xi^{u,x}_n)du
\newline \leq \lVert p''_\star\rVert_{\infty} \int_{\mathbb{R}} u^2 K(u) du .h^2_n 
\newline p_\star \text{ est de derivée seconde bornée donc } \lVert p''_\star\rVert_{\infty} \text{ existe, et } \int_{\mathbb{R}} u^2 K(u) du < \infty \text{. On pose } C_2 = \lVert p''_\star\rVert_{\infty} \int_{\mathbb{R}} u^2 K(u) du \text{, on a alors}
\newline biais(\widehat{p}^{h_n}_n(x)) \leq C_2.h^2_n
$$

$$
\mathcal{E}(x) = \mathbb{E}\left[\left(\widehat{p}_n(x)-p_{\star}(x)\right)^2\right]
\newline = Var(\widehat{p}_n(x)) + biais^2(\widehat{p}^{h_n}_n(x))
$$
D'après les résultats précédents, on a donc :
$$
\mathcal{E}(x) \leq C_1.(nh_n)^{-1} + C_2.h^4_n
\newline \text{ où } (C_1, C_2) \in \mathbb{R}^2
$$
On cherche alors à minimiser la fonction $h \longmapsto C_1.(nh)^{-1} + C_2.h^4 $
Le minimum de cette fonction, pour tout $n \in \mathbb{N}$ est proportionnel à $n^{-\frac{1}{5}}$.
$\newline$Pour fixer précisement $h_n$, on met en place un estimateur non biaisé de $\overline{\mathcal{E}}$, ce qui revient à estimer $J_n = \mathbb{E} \left[ \int_{\mathbb{X}} (\widehat{p}^{h_n}_n(x))^2 dx\right] -2 \int_{\mathbb{X}} p_\star(x) \mathbb{E} \left[ \widehat{p}^{h_n}_n(x) \right] dx$
$\newline$On met en place alors l'estimateur suivant:
$$\newline \widehat{J_n} = \int_{\mathbb{X}} (\widehat{p}^{h_n}_n(x))^2 dx - \frac{2}{n(n-1)h_n} \sum_{i=1}^n\sum_{\substack{j=1 \\ j≠i}}^n K \left( \frac{X_i - X_j}{h_n} \right)
$$
Finalement, on peut choisir $h_n$ de la façon suivante : 
$\newline h_n = c_\star.n^{-\frac{1}{5}} \text{ où } c_\star \in \underset{c}{\mathrm{argmin}} \, \widehat{J_n}
$


#### Code the theory

````python
def kernel_estimate(data, h, K, x):
    """
    Inputs
    ----------
    data: np.array contenant les X_i.
    h: fenêtre de l'estimateur.
    K: noyau utilisé
    
    Outputs
    -------
    kest: fonction retournant l'estimation de la densité pour chaque x donné en argument.
    """
    kest = 0
    n = data.shape[0]
    for i in range(n):
        X_i = data[i]
        kest += K((X_i-x)/h)
    kest /= (n*h)
    return kest
````

````python
import random
def K(u):
    return np.exp(-u**2/2)
###
# Tracé de l'estimateur pour différentes valeurs de h.
###
x = random.random()

fig = plt.figure(figsize=(12,12))
valeurs_de_h = np.arange(0.1, 20, 0.1)

estimateur_pour_h = [kernel_estimate(processed_data, h, K, x) for h in valeurs_de_h]
plt.scatter(valeurs_de_h, estimateur_pour_h)
plt.show()

sns.displot(estimateur_pour_h, kde = True)
plt.show()
````

On remarque que l'estimateur proposé dépend très fortement du choix de $h_n$ donc la valeur optimale dépend de constates inconnues. Une façon de choisir $h_n$ automatiquement est de minimiser un estimateur de l'erreur quadratique moyenne intégrée. Remarquons que 

$$
\overline{\mathcal{E}} = \mathbb{E}\left[\int_0^1\left(\widehat{p}^h_n(x)-p_{\star}(x)\right)^2 dx\right] = \int_0^1 p^2_{\star}(x)dx + \mathbb{E}\left[\int_0^1 (\widehat{p}^h_n(x))^2dx\right] - 2\int_0^1 p_{\star}(x)\mathbb{E}\left[\widehat{p}^h_n(x)\right]dx\,.
$$

Minimiser $\overline{\mathcal{E}}$ en fonction de $h$ est donc équivalent à minimiser $h \mapsto \mathbb{E}\left[\int_0^1 (\widehat{p}^h_n(x))^2dx\right] - 2\int_0^1 p_{\star}(x)\mathbb{E}\left[\widehat{p}^h_n(x)\right]dx$. Cette quantité étant inconnue, nous proposons de minimiser un estimateur sans biais de cette dernière.

### Proof that the estimator is unbiased

On note $\widehat{J_n}$ cet estimateur, ainsi $$\widehat{J_n} = \int_{\mathsf{X}} (\widehat{p}^{h}_n(x))^2 dx - \frac{2}{n(n-1)h} \sum_{i=1}^n\sum_{\substack{j=1 \\ j≠i}}^n K \left( \frac{X_i - X_j}{h} \right), \text{ pour tout h} \in \mathbb{R}$$
Montrons que cet estimateur est sans biais. Soit h $\in \mathbb{R}.$ $\newline$ 
Il suffit alors de montrer que $$\mathbb{E}\left[ \frac{1}{n(n-1)h}  \sum_{i=1}^n\sum_{\substack{j=1 \\ j≠i}}^n K \left( \frac{X_i - X_j}{h}\right) \right] = \int_0^1 p_{\star}(x)\mathbb{E}\left[\widehat{p}^h_n(x)\right]dx $$  
$\newline \text{Par indépendance, pour i} \neq \text{j, } (X_i, X_j) \text{ suit la densité } p_\star(x).p_\star(y) \text{, d'où : }$
$$
\mathbb{E}\left[ \frac{1}{n(n-1)h}  \sum_{i=1}^n\sum_{\substack{j=1 \\ j≠i}}^n K \left( \frac{X_i - X_j}{h}\right) \right] = \frac{1}{n(n-1)h} \sum_{i=1}^n\sum_{\substack{j=1 \\ j≠i}}^n \int_{\mathsf{X}^2} K\left( \frac{x-y}{h} \right) p_\star(x).p_\star(y) dxdy
\newline = \frac{1}{h} \int_{\mathsf{X}^2} K\left( \frac{x-y}{h} \right) p_\star(x).p_\star(y) dxdy
\newline = \int_0^1 p_\star(x) \left( \frac{1}{h} \int_0^1 K \left( \frac{x-y}{h} \right) p_\star(y) dy \right) dx
$$
$\text{Or, pour tout x} \in \mathsf{X},  $ $$\mathbb{E}\left[\widehat{p}^h_n(x)\right] = \mathbb{E}\left[ \frac{1}{nh_n}\sum_{i=1}^n K\left(\frac{X_i-x}{h_n}\right) \right] = \frac{1}{h} \int_0^1 K\left( \frac{z-x}{h} \right) p_\star(z)dz
$$
$ \text{Finalement, on peut simplifier l'équation précédente : } $
$$ 
\mathbb{E}\left[ \frac{1}{n(n-1)h}  \sum_{i=1}^n\sum_{\substack{j=1 \\ j≠i}}^n K \left( \frac{X_i - X_j}{h}\right) \right] = \int_0^1 p_\star(x) .\mathbb{E}\left[\widehat{p}^h_n(x)\right] dx
$$
$\text{L'estimateur } \widehat{J_n} \text{ est donc bien non biaisé.}$

````python
import random
import math
from scipy.integrate import quad 

def K(u):
    return np.exp(-u**2/2)

def integrated_MSE_estimate(data, h):
    """
    Inputs
    ----------
    data: np.array contenant les X_i.
    h: fenêtre de l'estimateur.
    
    Outputs
    -------
    integrated_MSE: estimateur dans biais défini en question Q4.
    """
    n = data.shape[0]
    
    x = random.random()
    sum_1 = 0
    for i in range(n):
        res, err = quad(K, (data[i]-1)/h, (data[i])/h) 
        sum_1 += res
    sum_1 /= (n**2*h**2)
    
    sum_11 = 0
    for i in range(n):
        for j in range(n):
            if i != j:
                res, err = quad(K, (-data[i]-data[j])/(2*h), (1-((data[i]+data[j])/2))/h)
                coeff = h/(2*np.pi)*np.exp(-data[i]/(2*h**2))*np.exp(-data[j]/(2*h**2))*np.exp((data[i]+data[j])**2/(4*h**2))
                sum_11 += coeff*res
    sum_11 /= (n**2*h**2)            
    
    sum_2 = 0
    for i in range(n):
        for j in range(n):
            if i != j:
                sum_2 += K((data[i] - data[j])/h)
    sum_2 *= 2/(n*(n-1)*h)
    
    integrated_MSE = sum_1 + sum_11 - sum_2
    return integrated_MSE
````

````python
def optimize_bandwith(data,range_h):
    """
    Inputs
    ----------
    data: np.array contenant les X_i.
    range_h: np.array contenant les valeurs possibles pour h
    
    Outputs
    -------
    h_opt: h optimal parmi range_h.
    """
    integrated_MSE_opti = integrated_MSE_estimate(data, range_h[0])
    h_opt = range_h[0]
    
    nb = 0
    for h in range_h:
        integrated_MSE = integrated_MSE_estimate(data, h)
        if integrated_MSE < integrated_MSE_opti:
            integrated_MSE_opti = integrated_MSE
            h_opt = h
        nb+=1
        print()
    
    return h_opt
````
